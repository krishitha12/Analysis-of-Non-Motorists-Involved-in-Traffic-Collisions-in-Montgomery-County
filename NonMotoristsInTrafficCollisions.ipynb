Function to convert Time from 12 hour format to 24 hour format.

Reference :https://www.geeksforgeeks.org/python-program-convert-time-12-hour-24-hour-format/

```python
def convert24(str1): 
      
    # Checking if last two elements of time 
    # is AM and first two elements are 12 
    if str1[-2:] == "AM" and str1[:2] == "12": 
        return "00" + str1[2:-2] 
          
    # remove the AM     
    elif str1[-2:] == "AM": 
        return str1[:-2] 
      
    # Checking if last two elements of time 
    # is PM and first two elements are 12    
    elif str1[-2:] == "PM" and str1[:2] == "12": 
        return str1[:-2] 
          
    else: 
          
        # add 12 to hours and remove PM 
        return str(int(str1[:2]) + 12) + str1[2:8]
```

converting Crash Date/Time column to 24 hour format and storing in a new column named Time In 24

```python
a=[]
for i in df['Crash Date/Time']:
    b=i.split(' ')
    a.append(b[0]+' '+convert24(b[1]+' '+b[2]))
df['Time In 24']=a
```

```python
df['Time In 24']=pd.to_datetime(df['Time In 24'].astype(str))
df
```

Function to return Hour from time given to it.

```python
def hr_func(ts):
    return ts.hour

df['Hours'] = df['Time In 24'].apply(hr_func)
```
